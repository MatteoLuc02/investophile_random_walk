```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_random_walk(n,k=1,prize=1,fees=0.001,p=0.5):
    for _ in range(k):
        array = np.random.random(n)
        res = []
        gain = 0
        for value in array:
            if value>p:
                gain+=prize-fees
            else:
                gain-=prize+fees
            res.append(gain)
        plt.plot(res)
    plt.show()
```
